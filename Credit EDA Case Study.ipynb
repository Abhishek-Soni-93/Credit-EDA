### Percentage of Missing Values

```python
def missing_values_table(application_df, print_info = False):
        # Total missing values
        mis_val = application_df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * application_df.isnull().sum() / len(application_df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        if print_info:
            # Print some summary information
            print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
                "There are " + str(mis_val_table_ren_columns.shape[0]) +
                  " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
```

### Treating outliers for above selected columns Using IQR Technique
#### Columns - (AMT_ANNUITY, AMT_CREDIT, AMT_INCOME_TOTAL)
From above three boxplot we can notice that there are significantly low amount of outliers present in these three columns. which are as follows:
    AMT_ANNUITY                2.44 %
    AMT_CREDIT                 2.13 %
    AMT_INCOME_TOTAL           4.56 %

So we will treat outliers using IQR technique.

IQR tells how spread the middle values are. It can be used to tell when a value is too far from the middle. From this technique an outlier is a point which falls more than 1.5 times the interquartile range above the third quartile or below the first quartile. Steps of IQR technique for utlier treatment are as below:

1. Arrange the data in increasing order
2. Calculate first(q1) and third quartile(q3)
3. Find interquartile range (q3-q1)
4. Find lower bound q1*1.5
5. Find upper bound q3*1.5
6. Anything that lies outside of lower and upper bound is an outlier
7. We will simply cap that outlier with upper bound and lower bound


```python
# Capping Outliers of column AMT_ANNUITY
AMT_ANNUITY_Q1 = application_df['AMT_ANNUITY'].quantile(0.25)
AMT_ANNUITY_Q3 = application_df['AMT_ANNUITY'].quantile(0.75)
AMT_ANNUITY_IQR = AMT_ANNUITY_Q3 - AMT_ANNUITY_Q1

# Calculating Lower and upper bound values of AMT_ANNUITY column
lower_bound_annuity = (AMT_ANNUITY_Q1 - 1.5 * AMT_ANNUITY_IQR) 
upper_bound_annuity = (AMT_ANNUITY_Q3 + 1.5 * AMT_ANNUITY_IQR) 

application_df['AMT_ANNUITY'].values[application_df['AMT_ANNUITY'].values > upper_bound_annuity] = upper_bound_annuity
application_df['AMT_ANNUITY'].values[application_df['AMT_ANNUITY'].values < lower_bound_annuity] = lower_bound_annuity
```

```python
# Capping Outliers of column AMT_CREDIT
AMT_CREDIT_Q1 = application_df['AMT_CREDIT'].quantile(0.25)
AMT_CREDIT_Q3 = application_df['AMT_CREDIT'].quantile(0.75)
AMT_CREDIT_IQR = AMT_CREDIT_Q3 - AMT_CREDIT_Q1

# Calculating Lower and upper bound values of AMT_CREDIT column
lower_bound_credit = (AMT_CREDIT_Q1 - 1.5 * AMT_CREDIT_IQR) 
upper_bound_credit = (AMT_CREDIT_Q3 + 1.5 * AMT_CREDIT_IQR) 

application_df['AMT_CREDIT'].values[application_df['AMT_CREDIT'].values > upper_bound_credit] = upper_bound_credit
application_df['AMT_CREDIT'].values[application_df['AMT_CREDIT'].values < lower_bound_credit] = lower_bound_credit
```

```python
# Capping Outliers of column AMT_INCOME_TOTAL
AMT_INCOME_Q1 = application_df['AMT_INCOME_TOTAL'].quantile(0.25)
AMT_INCOME_Q3 = application_df['AMT_INCOME_TOTAL'].quantile(0.75)
AMT_INCOME_IQR = AMT_INCOME_Q3 - AMT_INCOME_Q1

# Calculating Lower and upper bound values of AMT_INCOME_TOTAL column
lower_bound_income = (AMT_INCOME_Q1 - 1.5 * AMT_INCOME_IQR) 
upper_bound_income = (AMT_INCOME_Q3 + 1.5 * AMT_INCOME_IQR) 

application_df['AMT_INCOME_TOTAL'].values[application_df['AMT_INCOME_TOTAL'].values > upper_bound_income] = upper_bound_income
application_df['AMT_INCOME_TOTAL'].values[application_df['AMT_INCOME_TOTAL'].values < lower_bound_income] = lower_bound_income
```

```python
# Count of Outliers
Outliers_count = ((application_df < (Q1 - 1.5 * IQR)) | (application_df > (Q3 + 1.5 * IQR))).sum()
Outliers_count
```

```python
#Percent of missing values
def missing_values_table(Prev_df, print_info = False):
        # Total missing values
        mis_val = Prev_df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * Prev_df.isnull().sum() / len(Prev_df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        if print_info:
            # Print some summary information
            print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
                "There are " + str(mis_val_table_ren_columns.shape[0]) +
                  " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
```

### Bivariate Analysis

```python
def plot_data(feature,label_rotation=False,horizontal_layout=True):
    T = Merged_df[feature].value_counts()
    New_df = pd.DataFrame({feature: T.index,'Number of contracts': T.values})

    # Calculate the percentage of target=1 i.e. defaulter per category value
    Percentage_1 = Merged_df[[feature, 'TARGET']].groupby([feature],as_index=False).mean()
    Percentage_1.sort_values(by='TARGET', ascending=False, inplace=True)
    
    if(horizontal_layout):
        fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15,6))
    else:
        fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(12,14))
    sns.set_color_codes("pastel")
    s = sns.barplot(ax=ax1, x = feature, y="Number of contracts",data=New_df)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    
    s = sns.barplot(ax=ax2, x = feature, y='TARGET', order=Percentage_1[feature], data=Percentage_1)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    plt.ylabel('% of Defaulters (i.e. Target=1)', fontsize=12)
    plt.tick_params(axis='both', which='major', labelsize=12)

    plt.show()
```
